# Query Grants, Publications, Clinical Trials and patents

Using the Dimensions API, you can query Grants, Patents, and Clinical Trials as well as publications


## Grants

```python
from dimcli.shortcuts import dslquery_json as dslquery
import pandas as pd
```

```python
pd.DataFrame(

    dslquery("""
     search grants
     return grants
     limit 5
    """)['grants']
    
)
```

## Basic Funders Report

```python
def search_funders(limit,skip):
    searchstring = """search grants for "nanotechnology"
                  where start_year = 2016
               return funders aggregate funding sort by funding 
               limit {} 
             """.format(limit, skip)
    return searchstring
```

```python
skip = 0
funders = []
results = dslquery(search_funders(1000,skip))
funders = results['funders']

#while len(results['funders']) == 200:
#    skip += 200
#    results = dslquery(search_funders(200,skip))
#    funders += results['funders']
```

```python
pd.DataFrame(funders)
```

## Clinical Trials

```python
pd.DataFrame(
    
    dslquery("""
     search clinical_trials for "nanotechnology"
     return clinical_trials[all]
     limit 1000
    """)['clinical_trials']
    
)
```

## Patents

```python
list(pd.DataFrame(

dslquery("""
         search patents for "nanotechnology"
            
         return patents[all]
         limit 1
         skip 10
    """)['patents'] 

))
```
